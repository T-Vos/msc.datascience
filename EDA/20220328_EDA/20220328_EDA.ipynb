```python
import pandas as pd

def read_FH_DemBool():
    electoralDemocracyBool = pd.read_excel("../../../Data/FreedomHouse/List_of_Electoral_Democracies_FIW22.xlsx", sheet_name=0, header=1, engine='openpyxl')
    electoralDemocracyBool = electoralDemocracyBool.iloc[:, :2]
    d = {'Yes': True, 'No': False}
    electoralDemocracyBool['Electoral Democracy Designation in FIW 2022'] = electoralDemocracyBool['Electoral Democracy Designation in FIW 2022'].map(d)
    return electoralDemocracyBool

read_FH_DemBool()
```
