```python
import pandas as pd
import numpy as np
```

```python
def cut_year_title_down(title):
    # world bank columns end with [YRxxxx] EG: 1970 [YR1970]
    # This can be cut a way easily
    if title[0].isdigit() : title = title[0:4]
    return title
```

```python
def drop_non_year_column(titles):
    # Get Years in dataset
    drops = []
    for title in titles:
        if not title.isdigit() : drops.append(title)
    return drops
```

```python
educationData = pd.read_csv('../Data/WB_Edu_simple/646c5220-17f9-41f4-b39f-0f8c34130df6_Data.csv')
educationGDPData = pd.read_csv('../Data/Education_WB/edData.csv')
gdpPerCap = pd.read_csv('../Data/GDP2_WB/ad361792-9e4a-4b48-adb6-6422cf32b3fe_Data.csv')
tradeData = pd.read_stata('../Data/PennTrade/pwt100-trade-detail.dta')
tradeOfGDP = pd.read_csv('../Data/WB_trade/656d635f-d1ec-4345-a858-4784ff01e3f1_Data.csv')
taxesOfGDP = pd.read_csv('../Data/WB_taxesOfGDP/a77adba3-031c-495c-a100-610c369cdff7_Data.csv')
```

```python
gdpPerCap.columns = [cut_year_title_down(x) for x in gdpPerCap.columns]
educationData.columns = [cut_year_title_down(x) for x in educationData.columns]
educationGDPData.columns = [cut_year_title_down(x) for x in educationData.columns]
tradeOfGDP.columns = [cut_year_title_down(x) for x in tradeOfGDP.columns]
taxesOfGDP.columns = [cut_year_title_down(x) for x in taxesOfGDP.columns]
```

#### Create Dummy variable

```python
dataset["noGDPDummy"] = dataset["gdp"].isna()
dataset["noTaxDummy"] = dataset["taxOfGDP"].isna()
```

### Interpolation

```python
dataset["gdpLinInter"] = dataset.groupby('countryCode').apply(lambda group: group.interpolate())["gdp"]

dataset["taxOfGDPLinInter"] = dataset.groupby('countryCode').apply(lambda group: group.interpolate())["taxOfGDP"]

dataset["literaLinInter"] = dataset.groupby('countryCode').apply(lambda group: group.interpolate())["literarcy"]
dataset["eduLinInter"] = dataset.groupby('countryCode').apply(lambda group: group.interpolate())["eduOfGDP"]
dataset["primEducLinInter"] = dataset.groupby('countryCode').apply(lambda group: group.interpolate())["primOfGDP"]
dataset["secEducLinInter"] = dataset.groupby('countryCode').apply(lambda group: group.interpolate())["secOfGDP"]
dataset["tertEducLinInter"] = dataset.groupby('countryCode').apply(lambda group: group.interpolate())["tertOfGDP"]
```

### Log GDP

```python
from numpy import ma

dataset['gdpLinInterLog'] = ma.filled(np.log(ma.masked_equal(dataset['gdpLinInter'], 0)), 0)
```

### Create rolling mean

```python
dataset["rollingGDP10mean"] = dataset["gdpLinInter"].rolling(window=10).mean()
dataset['gdpLinInter10meanLog'] = ma.filled(np.log(ma.masked_equal(dataset['rollingGDP10mean'], 0)), 0)
dataset["rollingGDP5mean"] = dataset["gdpLinInter"].rolling(window=5).mean()
dataset['gdpLinInter5meanLog'] = ma.filled(np.log(ma.masked_equal(dataset['rollingGDP5mean'], 0)), 0)
dataset["rolling5YGDPStandardDev"] = dataset["gdpLinInter"].rolling(window=5).std()
dataset["gpd5YGrowth"] = dataset["gdpLinInter"].rolling(window=5).apply(lambda x: x.iloc[-1] - x.iloc[0])
dataset["rolling10YGDPStandardDev"] = dataset["gdpLinInter"].rolling(window=10).std()
dataset["gpd10YGrowth"] = dataset["gdpLinInter"].rolling(window=10).apply(lambda x: x.iloc[-1] - x.iloc[0])
dataset["gpdGrowth"] = dataset["gdpLinInter"].rolling(window=2).apply(lambda x: x.iloc[-1] - x.iloc[0])
dataset["gpdGrowth_stDev5Y"] = dataset["gpdGrowth"].rolling(window=5).std()
```
