```python
worldBankDict = dict(zip(worldBank['Country Name'], worldBank['Country Code']))
polityVDict = dict(zip(polity5Data['country'], polity5Data['scode']))
vDem = dict(zip(vDemDataSet['country_name'], vDemDataSet['country_text_id']))

auxillaryDictionary = {
    'Hong Kong':'HKG',
    "Bosnia":'BIH',
    'Burma/Myanmar':'MMR',
    'Egypt':'EGY',
    'Gambia':'GMB',
    'The Gambia':'GMB',
    'Democratic Republic of the Congo':'COD',
    'Iran':'IRN',
    'Korea North': 'PRK',
    'North Korea': 'PRK',
    'Korea South':'KOR',
    'South Korea': 'KOR',
    'Laos':'LAO',
    'Congo Brazzaville':'COG',
    'Congo Kinshasa':'COD',
    'Republic of the Congo':'COG',
    'Congo-Brazzaville': 'COG',
    "Cote D'Ivoire":'CIV',
    'Ivory Coast': 'CIV',
    'Yemen':'YEM',
    "United Kingdom":'GBR',
    "Cote D'Ivoire":'CIV',
    'Kyrgyzstan':'KGZ',
    'Myanmar (Burma)':'MMR',
    'Macedonia':'MKD',
    'Korea':'KOR',
    'Republic of Vietnam':'VNM',
    'Czechoslovakia':'CZE',
    'Cape Verde':'CPV',
    'Germany West':'DEU',
    'Russia':'RUS',
    'Syria':'SYR',
    'Slovakia':'SVK',
    'Slovak Republic':'SVK',
    'UAE':'ARE',
    'Timor Leste':'TLS',
    'United States                   ':'USA',
    'United States of America':'USA',
    'Venezuela':'VEN'
}
```

```python
totalCountryDictionary = {**polityVDict,**vDem, **auxillaryDictionary, **worldBankDict}
```

```python
polity5Data["baseLineCountryCode"] = polity5Data['country'].replace(totalCountryDictionary)
vDemDataSet["baseLineCountryCode"] = vDemDataSet['country_name'].replace(totalCountryDictionary)
```

```python
polity5Data["year"] = pd.to_numeric(polity5Data["year"])
```

```python
polity5Data = polity5Data[polity5Data["year"] >= 1960 & polity5Data["baseLineCountryCode"].isin(worldBankDict.values())]
```

```python
vDemDataSet = vDemDataSet[vDemDataSet["year"] >= 1960 & vDemDataSet["baseLineCountryCode"].isin(worldBankDict.values())]
```

```python
politySubsection = polity5Data[["democ","autoc",'polity','polity2', 'baseLineCountryCode','year']]
politySubsection["democ"] = politySubsection["democ"]/10
politySubsection["autoc"] = politySubsection["autoc"]/10
politySubsection["polity"] = (politySubsection["polity"]+10)/10
politySubsection["polity2"] = (politySubsection["polity2"]+10)/10
```

```python
politySubsection
```

```python
vDemSubsection = vDemDataSet[["v2x_polyarchy","v2x_libdem","v2x_partipdem","v2x_delibdem","v2x_egaldem","year","baseLineCountryCode"]]
vDemSubsection = vDemSubsection.groupby(["baseLineCountryCode","year"]).agg(['last']).reset_index()
```

```python
Y = politySubsection.merge(vDemSubsection,how="outer", sort=True)
```

```python
columnNames = {'baseLineCountryCode':'country'}
Y = Y.rename(columns=columnNames)
```

```python
Y.to_csv('y.csv', index=False)
```
